* given $$f(x_1,x_2)=4 x_1 + 6 x_2 - 2 x_1 ^ 2 - 2 x_1 x_2 - 2 x_2 ^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
f=lambda x,y: 4*x+6*y-2*x**2-2*x*y-2*y**2
x=np.linspace(-5,5,10)
y=x.copy()#creating 100 points in between -10 and 10
z=[]#creating an empty list to store the points generated for the plot
for i in range(0,len(x)):
    for j in range(i,len(y)):
        z.append(f(x[i],y[j]))
plt.plot(x,y,z)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

b. solving the problem using golden section method\
formula\
$x_1=x_r-\delta(x_r-x_l)$\
$x_2=x_l+\delta(x_r-x_l)$\
$if f(x_1)>f(x_2) then$\
$x_l=x_1$\
$x_r=x_r$\
$else if f(x_1)<f(x_2) then$\
$x_l=x_l$\
$x_r=x_2$\
$else if f(x1)=f(x2) then$\
$x_l=x_1$\
$x_r=x_2$\

```python
xl=30
xr=40
x1=0
x2=0
delta=(np.sqrt(5)-1)/2
#no of iterations=100
for i in range(0,100):
    x1=xr-delta*(xr-xl)
    x2=xl+delta*(xr-xl)
    if -f(xl,xr)>-f(x1,x2):
        xl=x1
        xr=xr
    elif -f(xl,xr)<-f(x1,x2):
        xl=xl
        xr=x2
    elif -f(xl,xr)==-f(x1,x2):
        xl=x1
        xr=x2
    print(f"at the {i}th iteration the value of x1 and x2 are{x1},{x2}")
    if x1-x2 == 0.0001:
        break#error rate assumed
print(f"therefor the function f(x,y) is maximum at the point {x1},{x2} and the value of the function at that point is {-f(x1,x2)}")
```
